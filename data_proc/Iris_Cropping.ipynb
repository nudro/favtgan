<b>Cropping Function</b>

```python
test_path = "/home/xxx/Documents/Data/test/examples/"
#dirs = os.listdir(path)
width, height = im.size #320, 220

def cropit(mypath):
    dirs = os.listdir(mypath)
    for item in dirs:
        fullpath = os.path.join(mypath, item)
        if os.path.isfile(fullpath):
            im = Image.open(fullpath)
            f, e = os.path.splitext(fullpath) #file and extension
            print(f)
            cropped = im.crop((25, 20, width-0, height-20)) #cropped
            cropped.save(f + '.bmp', "BMP", quality=100)
            display(cropped) #see in notebook

cropit(test_path)
```
