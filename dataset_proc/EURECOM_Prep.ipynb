````python
def tiff_to_tif(src_path, dst_path):
    for filename in glob.glob(os.path.join(src_path, '*.tiff')):
        tiff_file_name = os.path.basename(filename)[:-5]
        print(tiff_file_name)
        img = io.imread(filename)
        plt.imshow(img, cmap="gray")
        plt.axis("off")
        plt.savefig(dst_path + "/" + tiff_file_name + ".tif", transparent=True, dpi=300, bbox_inches="tight", pad_inches=0.0)
        
````

````python
def tif_to_jpg(src_path, dst_path):
    for filename in glob.glob(os.path.join(src_path, '*.tif')):
        tif_file_name = os.path.basename(filename)[:-4]
        print(tif_file_name)
        img = io.imread(filename)
        plt.imshow(img, cmap="gray")
        plt.axis("off")
        plt.savefig(dst_path + "/" + tif_file_name + ".jpg", transparent=True, dpi=300, bbox_inches="tight", pad_inches=0.0)
````

<h2>Resize and Pair</h2>

````python
neww, newh = 384, 256

def resize_unet256(mypath, outpath):
    #mypath: source (originals)
    #outpath: target (resized)
    dirs = os.listdir(mypath)
    counter = 1
    for item in dirs:
        fullpath = os.path.join(mypath, item)
        if os.path.isfile(fullpath):
            counter += 1
            im = Image.open(fullpath)
            f, e = os.path.splitext(fullpath) #file and extension
            resized = im.resize((neww, newh), Image.ANTIALIAS) #resize
            save_fname = os.path.join(outpath, os.path.basename(f) +'.bmp')
            resized.save(save_fname, quality=100)
            if counter <=10:
                display(resized) #see in notebook

````

````python
thermal = "eurecom/thermal_originals"
thermal_out = "eurecom/thermal_out"
resize_unet256(thermal, thermal_out)
````

````python
visible = "eurecom/visible_originals"
visible_out = "eurecom/visible_out"
resize_unet256(visible, visible_out)
````

<h2>How to pair</h2>

Create a directory structure like below. Make `A` be the visible and make `B` be the thermal. Under each, create a train folder using the Eurecom IDs in the data table on the README. 45 subjects go into training, the remaining 5 subjects: {1, 2, 21, 31, 36} go into the test set. You'll need to move these manually.

Now you are ready to pair the visible and thermal. We used the script at the official pix2pix repo: https://github.com/junyanz/pytorch-CycleGAN-and-pix2pix/blob/master/docs/datasets.md. 

Your directory should look like this: 

```
eurecom
- data
-- A
---train
---test
--B
---train
---test

- data_out # path where the combined datasets will live

```


```
Create folder /path/to/data with subfolders A and B. A and B should each have their own subfolders train, val, test, etc. In /path/to/data/A/train, put training images in style A. In /path/to/data/B/train, put the corresponding images in style B. Repeat same for other data splits (val, test, etc).

Corresponding images in a pair {A,B} must be the same size and have the same filename, e.g., /path/to/data/A/train/1.jpg is considered to correspond to /path/to/data/B/train/1.jpg.

Once the data is formatted this way, call:

python datasets/combine_A_and_B.py --fold_A /path/to/data/A --fold_B /path/to/data/B --fold_AB /path/to/data

```

This python script can be downloaded here: https://github.com/junyanz/pytorch-CycleGAN-and-pix2pix/blob/master/datasets/combine_A_and_B.py

Example usage: 

```
python datasets/combine_A_and_B.py --fold_A eurecom/data/A --fold_B eurecom/data/B --fold_AB eurecom/data_out

```
