<h2>EI_GANA</h2>

```python
from ir_EI_GANA_data import Faces
exp = 'iris_EI_GANA'
torchreid.data.register_image_dataset(exp, Faces)
```

```python
datamanager = torchreid.data.ImageDataManager(
    root='reid-data',
    sources=exp,
    targets=exp,
    height=256,
    width=128,
    batch_size_train=3,
    batch_size_test=3,
    transforms=['random_flip', 'random_crop']
)
model = torchreid.models.build_model(
    name='mudeep',
    num_classes=datamanager.num_train_pids,
    loss='softmax',
    pretrained=True
)

model = model.cuda()

optimizer = torchreid.optim.build_optimizer(
    model,
    optim='adam',
    lr=0.0003
)

scheduler = torchreid.optim.build_lr_scheduler(
    optimizer,
    lr_scheduler='single_step',
    stepsize=20
)
engine = torchreid.engine.ImageSoftmaxEngine(
    datamanager,
    model,
    optimizer=optimizer,
    scheduler=scheduler,
    label_smooth=True
)

engine.run(
    save_dir='log/mudeep',
    eval_freq=10,
    print_freq=10,
    test_only=True,
    visrank=True
)
```
