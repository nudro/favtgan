```python
import seaborn as sns
import pandas as pd
```

```python
def table_to_df(csv, exp):
    df = pd.read_csv(csv)
    df.columns = ['num_label', 'value', 'file']
    df['exp'] = exp
    return df
```

<h1>PSNR</h1>

```python
iris_pix2pixp = table_to_df('Iris/IR_GANB/psnr.csv', 'iris_pix2pix') #Iris pix2pix
EI_pix2pixp = table_to_df('Iris/EI_pix2pix/psnr.csv', 'EI_pix2pix') # EI pix2pix
EI_sensor_OGp = table_to_df('Iris/EI_sensor_OG/psnr.csv', 'EI_sensor_OG') #EI favtGAN baseline
EI_sensor_V3p = table_to_df('Iris/EI_sensor_V3/psnr.csv', 'EI_sensor_V3') #EI favtGAN no noise
EI_sensor_V4p = table_to_df('Iris/EI_sensor_V4/psnr.csv', 'EI_sensor_V4') #EI favtGAN noisy labels
EI_sensor_V5p = table_to_df('Iris/EI_sensor_V5/psnr.csv', 'EI_sensor_V5') #EI favtGAN Gaussian
IO_sensor_V4p = table_to_df('Iris/IO_sensor_V4/psnr.csv', 'IO_sensor_V4') #IO favtGAN noisy labels
```

```python
final_p = iris_pix2pixp.append([EI_pix2pixp,
                              EI_sensor_OGp,
                              EI_sensor_V3p,
                              EI_sensor_V4p,
                              EI_sensor_V5p,
                              IO_sensor_V4p])
final_p
```

```python
psnr_mean = pd.pivot_table(final_p,index=["exp"],values=["value"])
psnr_mean
```

```python
psnr_mean_by_subject = pd.pivot_table(final_p,index=["subject", "exp"],values=["value"], aggfunc=[np.mean])
psnr_mean_by_subject 
```

```python
def make_subjects(df):
    
    conditions = [
    (df['num_label'] >=0) & (df['num_label'] <= 31),
    (df['num_label'] >=32) & (df['num_label'] <= 64),
    (df['num_label'] >=65) & (df['num_label'] <= 97)]

    choices = ["Meng", "Vicki", "Vijay"]
    df['subject']= np.select(conditions, choices, default='null')

    return df
```

<h1>SSIM</h1>

```python
iris_pix2pixs = table_to_df('Iris/IR_GANB/ssim.csv', 'iris_pix2pix') #Iris pix2pix
EI_pix2pixs = table_to_df('Iris/EI_pix2pix/ssim.csv', 'EI_pix2pix')# EI pix2pix
EI_sensor_OGs = table_to_df('Iris/EI_sensor_OG/ssim.csv', 'EI_sensor_OG') #EI favtGAN baseline
EI_sensor_V3s = table_to_df('Iris/EI_sensor_V3/ssim.csv', 'EI_sensor_V3') #EI favtGAN no noise
EI_sensor_V4s = table_to_df('Iris/EI_sensor_V4/ssim.csv', 'EI_sensor_V4') #EI favtGAN noisy labels
EI_sensor_V5s = table_to_df('Iris/EI_sensor_V5/ssim.csv', 'EI_sensor_V5') #EI favtGAN Gaussian
IO_sensor_V4s = table_to_df('Iris/IO_sensor_V4/ssim.csv', 'IO_sensor_V4') #IO favtGAN noisy labels
```

```python
final_s = iris_pix2pixs.append([EI_pix2pixs,
                              EI_sensor_OGs,
                              EI_sensor_V3s,
                              EI_sensor_V4s,
                              EI_sensor_V5s,
                              IO_sensor_V4s])
final_s
```

```python
final_s = make_subjects(final_s)
```

```python
ssim_mean = pd.pivot_table(final_s,index=["exp"],values=["value"], aggfunc=[np.mean])
```

```python
ssim_mean_by_subject = pd.pivot_table(final_s,index=["subject", "exp"],values=["value"], aggfunc=[np.mean])
```
